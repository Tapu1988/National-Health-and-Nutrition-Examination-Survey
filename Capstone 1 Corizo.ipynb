# 1. On a single plot (use matplotlib.pytplot.subplot), draw two histograms: for female weights (top subfigure). and for male weights (bottom subfigure) Call matplotlib.pyplot.xlim to make the xaxis limits identical for both subfigures (work out the appropriate limits yourself).

```python
# To adjust the figure
plt.figure(figsize=(12,7))
# Female weights
female_weights = female[:, 0]

# Male weights
male_weights = male[:, 0]

# Create a figure with two subplots
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# Top subplot: Female weights histogram
axs[0].hist(female_weights, bins=20, alpha=0.5, label='Female')
axs[0].set_title('Female Weights')
axs[0].set_xlabel('Weight (kg)')
axs[0].set_ylabel('Frequency')

# Bottom subplot: Male weights histogram
axs[1].hist(male_weights, bins=20, alpha=0.5, label='Male')
axs[1].set_title('Male Weights')
axs[1].set_xlabel('Weight (kg)')
axs[1].set_ylabel('Frequency')

# Set identical x-axis limits for both subplots
xmin = min(np.min(female_weights), np.min(male_weights))
xmax = max(np.max(female_weights), np.max(male_weights))
plt.xlim(xmin, xmax)

# Show the plot
plt.tight_layout()
plt.show()
```

## Inference
1. The female weight distribution is more concentrated and has a smaller range (approximately 40-80 kg) compared to the male weight distribution (approximately 50-120 kg).
2. The male weight distribution is more spread out and has a longer tail on the right side, indicating more extreme values.
3. The median weight for females is around 60 kg, while for males it is around 80 kg.
4. Both distributions are slightly skewed to the right, indicating more extreme values on the higher end of the weight range.

# 3. Compute the basic numerical aggregates of the male and female weights (measures of location, dispersion, and shape). In your own words, describe and compare the two distributions (e.g., are they left skewed, which one has more dispersion, and so forth).

```python
# Measures of location
female_mean = np.mean(female_weights)
male_mean = np.mean(male_weights)
female_median = np.median(female_weights)
male_median = np.median(male_weights)

# Measures of dispersion
female_std = np.std(female_weights)
male_std = np.std(male_weights)
female_var = np.var(female_weights)
male_var = np.var(male_weights)

# Measures of shape
female_skew = np.mean((female_weights - female_mean) ** 3) / (female_std ** 3)
male_skew = np.mean((male_weights - male_mean) ** 3) / (male_std ** 3)
female_kurt = np.mean((female_weights - female_mean) ** 4) / (female_std ** 4)
male_kurt = np.mean((male_weights - male_mean) ** 4) / (male_std ** 4)

print("Female Weights:")
print("Mean:", female_mean)
print("Median:", female_median)
print("Standard Deviation:", female_std)
print("Variance:", female_var)
print("Skewness:", female_skew)
print("Kurtosis:", female_kurt)

print("\nMale Weights:")
print("Mean:", male_mean)
print("Median:", male_median)
print("Standard Deviation:", male_std)
print("Variance:", male_var)
print("Skewness:", male_skew)
print("Kurtosis:", male_kurt)
```

# Inference:

1. The mean and median weights for males are higher than those for females.
2. The standard deviation and variance of male weights are greater than those for female weights, indicating more dispersion in the male weight distribution.
3. Both distributions are slightly right-skewed (positive skewness), but the male weight distribution is more skewed.
4. The kurtosis values indicate that both distributions have heavier tails than a normal distribution, but the male weight distribution has a more extreme tail.

# 5. Create a new matrix zfemale being a version of the female dataset with all its columns standardised (by computing the z-scores of each column).

```python
# Standardize the female dataset
zfemale = (female1 - np.mean(female1, axis=0)) / np.std(female1, axis=0)

print(zfemale)
```

# Inference:

1. The zfemale matrix contains the standardized values of the original female dataset.
2. Each column of zfemale has a mean of 0 and a standard deviation of 1.
3. The standardized values are also known as z-scores, which indicate how many standard deviations away from the mean each value is.
4. The axis=0 argument in np.mean and np.std calculates the mean and standard deviation for each column separately.

# 9. In your own words, list some advantages and disadvantages of BMI, waist-to-height ratio, and waist-to-hip ratio.

# Here are the advantages and disadvantages of BMI, waist-to-height ratio, and waist-to-hip ratio in my own words:

## BMI (Body Mass Index)

### Advantages:

1. Easy to calculate: BMI is simple to calculate using height and weight measurements.
2. Widely used: BMI is widely used by healthcare professionals and researchers.
3. Quick assessment: BMI provides a quick assessment of weight status.

### Disadvantages:

1. Limited accuracy: BMI does not account for muscle mass or body composition.
2. Misclassification: BMI can misclassify athletes or individuals with high muscle mass as overweight.
3. Ignores central obesity: BMI does not account for central obesity, which is a risk factor for chronic diseases.

## Waist-to-Height Ratio

### Advantages:

1. Accounts for central obesity: Waist-to-height ratio accounts for central obesity, which is a risk factor for chronic diseases.
2. Simple to calculate: Waist-to-height ratio is easy to calculate using waist circumference and height measurements.
3. Better predictor: Waist-to-height ratio is a better predictor of health risks compared to BMI.

### Disadvantages:

1. Limited research: There is limited research on waist-to-height ratio compared to BMI.
2. Variability: Waist-to-height ratio can vary depending on age, sex, and ethnicity.

## Waist-to-Hip Ratio

### Advantages:

1. Accounts for body shape: Waist-to-hip ratio accounts for body shape and central obesity.
2. Predictor of health risks: Waist-to-hip ratio is a predictor of health risks such as cardiovascular disease and diabetes.
3. Accounts for muscle mass: Waist-to-hip ratio accounts for muscle mass and body composition.

### Disadvantages:

1. More complex calculation: Waist-to-hip ratio requires waist and hip circumference measurements, making it more complex to calculate.
2. Limited use: Waist-to-hip ratio is not as widely used as BMI or waist-to-height ratio.
3. Variability: Waist-to-hip ratio can vary depending on age, sex, and ethnicity.




# 10. Print out the standardised body measurements for the 5 persons with the lowest BMI and the 5 persons with the 5 highest BMI (e.g., call print for a subset of zfemale comprised of 10 chosen rows as determined by a call to numpy.argsort). Interpret the results.

```python
# Get indices of 5 persons with lowest and highest BMI
lowest_bmi_indices = np.argsort(female_bmis)[:5]
highest_bmi_indices = np.argsort(-female_bmis)[:5]

# Combine indices
selected_indices = np.concatenate((lowest_bmi_indices, highest_bmi_indices))

# Print standardized body measurements for selected persons
print("Selected Body Measurements:")
print("Weight(kg),Height(cm),UpperArmLength(cm),UpperLegLength(cm),ArmCircumference(cm),HipCircumference(cm),Waist Circumference(cm)")
print("Lowest BMI:")
print(female[lowest_bmi_indices])
print("\nHighest BMI:")
print(female[highest_bmi_indices])

# Print BMI for selected persons
print("\nSelected BMI:")
print(np.round(female_bmis[selected_indices], 2))
```

# Interpretation:


1. The 5 persons with the lowest BMI have relatively low weights and body circumferences.
2. The 5 persons with the highest BMI have relatively high weights and body circumferences.
3. There is a significant difference in BMI between the lowest and highest groups.
4. The results suggest that BMI is strongly correlated with body weight and circumference.
